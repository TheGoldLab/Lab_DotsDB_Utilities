Below, produce and display as a movie a single trial on the fly.

```python
default_params = dict(
        speed=2.1,
        density=90,
        coh_mean=50,
        coh_stdev=10,
        direction='left',
        num_frames=30,
        diameter=10,
#         pixels_per_degree=55,
#         dot_size_in_pxs=1
    )

stim = DotsStimulus(**default_params)

normalized_frames = list(stim.normalized_dots_frame_generator())
px_frames = [stim.norm_to_pixel_frame(f) for f in normalized_frames]
movie = np.stack([f.transpose() for f in px_frames], axis=-1)

kstim.play_movie(movie, framerate=1/stim.frame_rate)
```

Now, I will display 3 random trials from an existing dataset.

```python
import h5py
```

```python
def format_trial_from_db_as_movie(tn):
    f = h5py.File('test.h5', 'r')
    g = f['intlv3_lftTrue_fr60_fs1.1_sp2.1_ds90_c50_cs10_dleft_nf6_dm10_sc5.0_ppd55.4612_dts4_fw610']
    s = g['px']
    npx = g.attrs['frame_width_in_pxs']
    nf = g.attrs['num_frames']
    fr = g.attrs['frame_rate']
    trial = s[tn-1,:]
    f.close()
    return trial.reshape((npx, npx, nf), order="F")
```

```python
kstim.play_movie(format_trial_from_db_as_movie(199), framerate=1/60)
```
